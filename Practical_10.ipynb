```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

data = p.read_csv("Iris.csv")
```

```python
def graph(y):
    sns.boxplot(x="variety",y=y,data=data)

plt.figure(figsize=(10,10))

plt.subplot(221)
graph('sepal.length')

plt.subplot(222)
graph('sepal.width')

plt.subplot(223)
graph('petal.length')

plt.subplot(224)
graph('petal.width')

plt.show()
```
